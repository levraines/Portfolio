```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0,5.0,0.1)
x2 = np.arange(0,5.0,0.2)
```

```python
plt.figure(1) # este parametro es por defecto

plt.subplot(211) # indica el numero de filas y columnas, y se utilizan dos filas y una columna, el 1 indica el primer dibujo
plt.plot(x1,f(x1), 'ro', x2, f(x2), 'k')

plt.subplot(212) # se utilizan dos filas y una columna, y el 2 es el segundo grafico
plt.plot(x2,f(x2), 'g--')

plt.show()
```

### Textos en los graficos

```python
mu = 100
sigma = 20

x = mu + sigma * np.random.randn(10000)
```

```python
n, bins, patches = plt.hist(x, 50, normed=1, facecolor = "g", alpha = 0.6) # normed seria para ver el eje y en terminos relativos
plt.xlabel("Coeficiente Intelectual", fontsize = 14, color = 'g')
plt.ylabel("Probabilidad")
plt.title(r"Histograma de CI $N(\mu, \sigma)$") # podemos agregar notacion de Latex
plt.text(120, 0.015, r'$\mu = 100,\ \sigma=20$') # localizacion en el grafico para agregar el texto
plt.axis([20,180, 0,0.025])
plt.grid(True) # cuadricula para poder leer los datos 
plt.show()
```

```python
plt.figure(figsize = (10,6), dpi = 90) # parametros para definir la calidad del grafico
plt.subplot(1,1,1)

x = np.arange(0, 10*np.pi, 0.01)
y = np.cos(x)
plt.plot(x, y, lw = 2.0) # lw es linewidth

plt.annotate("Maximo Local", xy = (4*np.pi, 1), xytext = (15, 1.5), 
             arrowprops = dict(facecolor = 'black', shrink = 0.05)) # tamano de la flecha
plt.ylim(-2,2)
plt.show()
```

```python
from matplotlib.ticker import NullFormatter
```

```python
mu = 0.5
sd = 0.3
y = mu + sd*np.random.randn(1000)
y = y[(y>0) & (y<1)]
y.sort()
x = np.arange(len(y))
```

```python
plt.figure(figsize = (10,8))

plt.subplot(2,2,1)
plt.plot(x,y)
plt.yscale("linear") # escala lineal 
plt.xscale("linear")
plt.title("Escala Lineal")
plt.grid(True) # para ver la cuadricula

plt.subplot(2,2,2)
plt.plot(x,y)
plt.yscale("log") # escala logaritmica
plt.title("Escala Logaritmica")
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(x,y-y.mean())
plt.yscale("symlog", linthreshy = 0.01 ) # escala logaritmica simetrica
plt.title("Escala Log Simetrico")
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale("logit") # escala logistica
plt.title("Escala Logistica")
plt.gca().yaxis.set_minor_formatter(NullFormatter())
plt.grid(True)

plt.subplots_adjust(top = 0.92, bottom = 0.08, left= 0.10, right = 0.95, hspace = 0.35, wspace = 0.35) # evitar la superposicion de los graficos

plt.show()
```

### Cambios en los ejes

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True) # sino se coloca endpoint el ultimo digito no se genera
S,C = np.sin(x), np.cos(x)
```

```python
plt.plot(x,S, color = 'b', lw = 1.2, linestyle = "-")
plt.plot(x,C, color = "g", lw = 1.2, linestyle = "-")
plt.xlim(x.min()*1.1, x.max()*1.1) # para cambiar los ejes en x
plt.ylim(S.min()*1.1, S.max()*1.1) # para cambiar los ejes en y
plt.xticks(np.linspace(-4,4,17, endpoint = True))
plt.yticks(np.linspace(-1,1,9, endpoint = True)) # para cambiar el tamano de los ejes va de -1 a 1 y con 9 divisiones
plt.show()
```

```python
plt.figure(figsize = (10,8))

plt.plot(x,S, color = 'b', lw = 1.2, linestyle = "-", label = "seno")
plt.plot(x,C, color = "g", lw = 1.2, linestyle = "-", label = "coseno")
plt.xlim(x.min()*1.1, x.max()*1.1) # para cambiar los ejes en x
plt.ylim(S.min()*1.1, S.max()*1.1) # para cambiar los ejes en y
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$]', r'$0$', r'$+\pi/2$', r'+$\pi$'])
plt.yticks(np.linspace(-1,1,2, endpoint = True),
           ['-1', '+1']) # para cambiar el tamano de los ejes va de -1 a 1 y con 9 divisiones

ax = plt.gca()
ax.spines['right'].set_color('none') # para quitar las lineas de; cuadrado 
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

plt.legend(loc = 'upper left') # posicion donde queremos poner la caja de leyendas

x0 = 2*np.pi/3
plt.plot([x0,x0],[0,np.sin(x0)], color = "blue", lw = 2.5, linestyle = "--") # para hacer una division en el seno 
plt.scatter([x0, ], [np.sin(x0), ], 50, color = "blue")
plt.annotate(r'$\sin(\frac{2\pi}{3}) = \frac{\sqrt{3}}{2}$', 
            xy = (x0, np.sin(x0)), xycoords = "data", 
            xytext = (+20,+40), textcoords = "offset points",
            fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle= "arc3,rad =.2"))
x0 = 2*np.pi/3
plt.plot([x0,x0],[0,np.cos(x0)], color = "green", lw = 2.5, linestyle = "--") # para hacer una division en el seno 
plt.scatter([x0, ], [np.cos(x0), ], 50, color = "green")
plt.annotate(r'$\sin(\frac{2\pi}{3}) = -\frac{1}{2}$', 
            xy = (x0, np.cos(x0)), xycoords = "data", 
            xytext = (-90,-60), textcoords = "offset points",
            fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle= "arc3,rad =.2"))

for label in ax.get_xticklabels()+ ax.get_yticklabels(): # bucle para cambiar tamano de las etiquetas 
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor="cyan", edgecolor = "None", alpha = 0.6)) 


plt.show()
```
