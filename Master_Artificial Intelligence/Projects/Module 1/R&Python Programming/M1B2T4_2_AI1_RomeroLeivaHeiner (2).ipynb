Por su parte,  Hayden Christensen (actor que interpretó a Anakin) medía solo 6 pies, así se explica a nivel de la vida real. 

2. **En el universo de StarWars**: probablemente tenga algo que ver con el hecho de que Vader literalmente tiene nuevos pies. Después de la pelea en Mustafar, Vader obtuvo nuevas partes que no coincidían exactamente con las que perdió. Claramente, eran más grandes y agregaban más de unas pocas pulgadas.

## Ejercicio 1.5.

Calcula, a partir de los vectores numpy de altura y peso, definidos a continuación, el [índice de masa corporal (IMC)](https://es.wikipedia.org/wiki/%C3%8Dndice_de_masa_corporal) de los personajes de star wars contenidos en people_df: 

$IMC = \frac{peso}{altura^{2}}$ donde altura está en metros y el peso en kg.

Muestra los datos por pantalla.

```python
# Solución:

IMC = people_df.mass/(people_df.height/100)**2
IMC
```

## Ejercicio 1.6.

A partir del IMC que has calculado en el ejercicio 1.5. Construye un panel con dos histogramas:

- Un histograma con toda la muestra
- Un histograma seleccionando los valores con un IMC inferior a 100.

```python
# Solución:
sns.set()
plt.figure();  # crea la figura del plot

# crea el primer panel con toda la muestra
plt.subplot(2, 1, 1) # (filas, columnas, número de panel)
plt.hist(IMC, color="turquoise", label="Indice de Masa Corporal")
plt.xlabel("IMC",labelpad=-6)  # labelpad desplaça el titulo arriba
plt.ylabel("N")
plt.legend();

# crea el segundo panel solo con individuos que tengan un IMC < 100
plt.subplot(2, 1, 2)
plt.hist(IMC[IMC<100], color="deepskyblue", label="Indice de Masa Corporal < 100");
plt.xlabel("IMC")
plt.ylabel("N")
plt.legend();
```

## Ejercicio 1.7.

A partir del vector 1 y 2 que se definen a continuación contesta las siguientes preguntas:

- Calcula el shape, ndim, size del vector1 y vector2
- Explica cual es la diferencia entre vector1 y vector2 a partir de los que hayas observado

```python
altura = people_df.height
peso = people_df.mass

vector1 = np.hstack([altura,peso])
vector2 = np.vstack([altura,peso])
```

```python
# Solución:

print(vector1.shape, vector1.ndim, vector1.size)
print(vector2.shape, vector2.ndim, vector2.size)
```

### Analisis

Con la función **hstack** aplicada al vector1 apila la secuencia de matrices de entrada horizontalmente (es decir, en columnas) para hacer una sola matriz, de ahí que la altura que tiene 36 elementos y peso que cuenta con la misma cantidad de elementos, una vez aplicado el hstack se ve que el nuevo shape es de 72 elementos y de una sola dimensión. 

Con la función **vstack** aplicada al vector2 apila la secuencia de matrices de entrada verticalmente (es decir, en filas) de ahí que el altura y peso, que tienen la misma cantidad de elementos (36) ahora es una matriz de dos dimensiones y de 36 elementos en cada dimensión.
