**La probabilidad de sobrevivir en cualquiera de los seis países europeos seleccionados para el estudio.**

```python
import warnings
warnings.filterwarnings("ignore")

data['P_supervivencia'] = round(100 - (data['Total Deaths']/data['Total Cases'])*100, 3)
```

```python
# Visualizando nuevo dataset
data.head(5)
```

```python
# Función programada para mostrar probabilidad de supervivencia
def show_values(axs, orient="v", space=.02):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height()*0.02)
                value = '{:.2f}'.format(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.5)
                value = '{:.2f}'.format(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```

```python
plt.figure(figsize = (15,8))
plt.title('Probabilidad de Sobrevivir', y=1, fontsize = 20)
plt.suptitle('(Valores expresados en porcentaje)', y=0.87, fontsize = 12)
sns.set_theme(style="whitegrid")
ax = sns.barplot(data = data, x="Country", y="P_supervivencia")
show_values(ax)
```

**Análisis:** Se puede observar como en el país que más probabilidades hay de sobrevivir es en Ucrania, seguido de Alemania. En el que menos probabilidad hay de sobrevivir es en Italia. Aquí el rol de un analista de datos puede ser crucial ya que nos podemos preguntar, ¿Por qué Italia es el país con la menor tasa de supervivencia en comparación con los otros países? Y aquí se puede empezar a mirar diversas aristas como: políticas públicas de salud, calidad de los centros de atención sanitarios, medidas restrictivas para ralentizar la propagación, etc.

**Si seleccionamos una de las personas fallecidas, ¿cuál es la probabilidad de que residiera en España? ¿Y en el resto de países?**



```python
# Creando nueva variable de probabilidad de fallecimiento
data['P_fallecimiento'] = round((data['Total Deaths'] / sum(data['Total Deaths'])*100), 3)
data
```

```python
plt.figure(figsize = (15,8))
#plt.title('Probabilidad de fallecimiento', fontsize = (20))
plt.title('Probabilidad de fallecimiento', y=1, fontsize = 20)
plt.suptitle('(Valores expresados en porcentaje)', y=0.87, fontsize = 12)

sns.set_theme(style="whitegrid")
ax = sns.barplot(data = data, x="Country", y="P_fallecimiento")
show_values(ax)
```

**Análisis:** se puede observar que en el país que hay más probabilidad de perecer es en UK (ya habíamos apuntado a esto en el EDA), seguido de Italia y después Francia. En el país que menos probabilidad hay de perecer es en Ucrania, seguido de Alemania con solo 1.70% y 6.31% respectivamente, aquí también se pueden hacer varias preguntas como: ¿Qué están haciendo estos países para tratar a las personas infectadas? ¿Cómo es su sistema de salud? ¿Cómo es su sistema de atención primaria y de prevención sanitario?

**Conclusiones finales:**

A nivel general tomando en cuenta únicamente los 6 países estudiados se observa que en países con una cantidad de población total similar, como es el caso de España y Ucrania, la probabilida de supervivencia es diferente y habría que hacer un análisis integral para entender el porqué. Se pudo comprobar como el país con la mayor tasa de supervivencia es Ucrania, seguido de Alemania y el que presenta la menor tasa de supervivencia es Italia. 

Por otro lado el país que presenta la mayor tasa de mortalidad es UK seguido de Italia, mientras que los que presentan las más bajas son Ucrania seguido de Alemania.
