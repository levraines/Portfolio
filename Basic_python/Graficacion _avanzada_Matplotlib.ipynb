#### 1. A partir del siguiente código, intente obtener el gráfico del dibujo llamado ejercicio 1 en el enunciado.

Pista: investigue el parámetro fill_between de matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 256 # dado por profesor
X = np.linspace(-np.pi, np.pi, n, endpoint=True) # dado por profesor
Y = np.sin(2 * X) # dado por profesor

plt.axes([0.025, 0.025, 0.95, 0.95]) # tamano del grafico
plt.plot(X, Y + 1, color='blue', alpha=1.00) # dado por profesor
plt.fill_between(X, 1, Y + 1, color='blue', alpha=.25) # para primer grafico
plt.plot(X, Y - 1, color='blue', alpha=1.00) # dado por profesor
plt.fill_between(X, -1, Y - 1, (Y - 1) > -1, color='blue', alpha=.25) # segunda grafico > 1
plt.fill_between(X, -1, Y - 1, (Y - 1) < -1, color='red',  alpha=.25) # segunda grafico < 1
plt.xlim(-np.pi, np.pi) # limite
plt.xticks(())  # vacia para que no dibuje los puntos 
plt.ylim(-2.5, 2.5) #limite y 
plt.yticks(()) # vacia para que no dibuje los puntos 
plt.show() # muestra el grafico
```

#### 2. A partir del siguiente código, intente obtener el gráfico del dibujo llamado ejercicio 2 en el enunciado.

Pista: el color viene dado por el ángulo que forman x e y con la horizontal

```python
n = 1024 #dado por profesor
X = np.random.normal(0, 1, n) #dado por profesor
Y = np.random.normal(0, 1, n) #dado por profesor
T = np.arctan2(Y, X) # angulo que se forma y que define el color

plt.axes([0.025, 0.025, 0.95, 0.95]) # tamano
plt.scatter(X, Y, s=75, c=T, alpha=.5) # C color definido por el arcotangente y el alpha opacidad
plt.xlim(-1.5, 1.5) # limite 
plt.xticks(()) # esconde numereracion en el eje x
plt.ylim(-1.5, 1.5) # limite y
plt.yticks(()) # esconde numereracion en el eje x

plt.show() # muestra el grafico
```

```python
n = 12 #dado por profesor / cantidad de bloques que se desean obtener 
X = np.arange(n) #dado por profesor / el rango obtenido
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n) #dado por profesor
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n) #dado por profesor

plt.axes([0.025, 0.025, 0.99, 0.99]) # tamano de la figura
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white') #dado por profesor
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white') #dado por profesor

for x, y in zip(X, Y1):
    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='right', va= 'bottom') #dado por profesor / ha = alineacion de etiquetas 


for x, y in zip(X, Y2):
    plt.text(x + 0.4, -y - 0.05, '%.2f' % y, ha='right', va= 'top') # llamo un for parecido al dado y cambio parametros


plt.xlim(-0.5, n) # limite de x
plt.xticks(()) # quita numeros 
plt.ylim(-1.25, 1.25) #dado por profesor, # limites de Y, van desde -1.25 a 1.25
plt.yticks(()) # quita numeros

plt.show() # muestra grafico
```

```python
n = 20 # cantidad de partes del grafico de pastel 
Z = np.ones(n) #creo 20 partes 
Z[-1] *= 2  # multiplico el -1 por 2 para obtener la cantidad de segmentos 

plt.axes([0.025, 0.025, 0.95, 0.95]) # tamano
plt.pie(Z, explode=Z*.05, colors = ['%f' % (i/float(n)) for i in range(n)])
plt.axis('equal')
plt.xticks(())
plt.yticks()

plt.show()
```
