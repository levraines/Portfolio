## Diagramas de caja y bigotes

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(19880519)
```

```python
dist1 = 100*np.random.rand(50)
dist2 = 50*np.ones(25)
dist3 = 100+100*np.random.rand(10)
dist4 = -100*np.random.rand(10)
data = np.concatenate((dist1, dist2, dist3, dist4))
```

```python
fig1, ax1 = plt.subplots()
ax1.set_title("Boxplot basico")
ax1.boxplot(data)
plt.show()
```

```python
fig2, ax2 = plt.subplots()
ax2.set_title("Boxplot con muescas")
ax2.boxplot(data, notch = True)
plt.show()
```

```python
green_diamond = dict(markerfacecolor = "green", marker = "D")
fig3, ax3 = plt.subplots()
ax3.set_title("Boxplot con outliers personalizados")
ax3.boxplot(data, flierprops = green_diamond)
plt.show()
```

```python
fig4, ax4 = plt.subplots()
ax4.set_title("Boxplot sin outliers")
ax4.boxplot(data, showfliers = False) # para quitar los outliers
plt.show()
```

```python
red_square = dict(markerfacecolor = "r", marker = "s")
fig5, ax5 = plt.subplots()
ax5.set_title("Boxplot personalizado y en horizontal")
ax5.boxplot(data, vert = False, flierprops = red_square, whis = 1.5) # para quitar los outliers
plt.show()
```

```python
dist1 = 100*np.random.rand(50)
dist2 = 40*np.ones(25)
dist3 = 100+100*np.random.rand(10)
dist4 = -100*np.random.rand(10)
data2 = np.concatenate((dist1, dist2, dist3, dist4))
```

```python
data.shape = (-1,1)
data2.shape = (-1,1)
```

```python
fulldata = [data, data2, data2[::2,0]] # datos combinados 
```

```python
fig7, ax7 = plt.subplots()
ax7.set_title("Multiples muestras de diferentes naturaleza")
ax7.boxplot(fulldata)
plt.show()
```
